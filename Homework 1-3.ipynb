#### Q1: <b>(8 marks)</b> Activation Functions 

Following are some common activation functions implemented using Python. 

```python

    return x
```

```python
def binaryStep(x):
    return np.heaviside(x,1)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def tanh(x):
    return np.tanh(x)
```   

```python
def relu(X):
   return np.maximum(0,X)
```

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

````python
#Define the Activation Functions Here
#Define the Activation Functions Here
def linear(x):
    return x

def binaryStep(x):
    return np.heaviside(x,1)

def sigmoid(x):
    return 1/(1+np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(X):
    return np.maximum(0,X)

def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)
````

<b>Q12. (12 marks) For the neural network from Q11 given below, generate three test cases with both X and y.  y should be one hot encoded. For example, [0,1,0] means that the correct answer is the second node (e.g., 2).</b>

![title](H1F11.png)


<ol>
<li>Write a function <i>predict(x)</i> that takes any input for this neural network and uses the weights and activation functions to return the predicted value. </li>
<li>Write a function <i> mse(x,y) </i> that returns the mean-squared error between two vectors x and y.</li>
<li>Write a function <i> diff(x,y) </i> that returns a 1 if the node chosen by the neural network (e.g., 2) is the same as that indicated by the respective y (e.g., [0,1,0]), otherwise it returns 0.</li>
<li>Calculate and print the mean square error (mse) for each of the test cases using the two functions defined above. </li>
<li>Calculate and print the diff for each of the test cases using the two functions defined above. </li>
<li>Calculate and print the percentage of answer the neural network got right using the two functions defined above. </li>
</ol>

````python
# Ans for Q12
# Function to generate test cases with one hot encoded y
def generate_test_cases(num_cases):
    test_cases = [(np.random.random(3), np.eye(3)[np.random.choice(3)]) for _ in range(num_cases)]
    return test_cases

# Function to predict output using neural network
def predict(x, w1, w2, w3):
    return forward(x, w1, w2, w3)

# Function to calculate mean squared error between two vectors
def mse(x, y):
    return np.mean((x - y) ** 2)

# Function to check if the predicted node is the same as the actual node
def diff(x, y):
    return 1 if np.argmax(x) == np.argmax(y) else 0

# Initialize random weights
W1 = np.random.random((1,3))
W2 = np.random.random((2,1))
W3 = np.random.random((3,2))

# Generate test cases
test_cases = generate_test_cases(3)

# Calculate and print mse, diff and percentage of correct answers
correct_count = 0
for i, (x, y) in enumerate(test_cases):
    predicted_y = predict(x, W1, W2, W3)
    print(f"Test Case {i+1}:\nX = {x}\nY = {y}\nPredicted Y = {predicted_y}")
    print(f"MSE: {mse(predicted_y, y)}")
    result = diff(predicted_y, y)
    correct_count += result
    print(f"Diff: {result}")

# Calculate percentage of correct answers
percentage_correct = (correct_count / len(test_cases)) * 100
print(f"Percentage of correct answers: {percentage_correct}%")
````
